```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/final_dataset updated (3).csv')
```

```python
df.sample(10)
# df.drop(df[(pd.to_datetime(df['date']).dt.month == 1) & (pd.to_datetime(df['date']).dt.day == 1)].index, inplace=True)
df.drop(df[(pd.to_datetime(df['date']).dt.month == 1) & (pd.to_datetime(df['date']).dt.day == 1)].index, inplace=True)
```

```python
df.info()
```

```python
plt.figure(figsize=(10, 8))
plt.subplot(2, 2, 1)
ax = sns.histplot(df['income'], kde=True)
ax.lines[0].set_color('crimson')
plt.title('Income')

plt.subplot(2, 2, 2)
ax = sns.histplot(df['expense'], kde=True)
ax.lines[0].set_color('crimson')
plt.title('Expenses')

plt.subplot(2, 2, 3)
ax = sns.histplot(df['emergency_fund'], kde=True)
ax.lines[0].set_color('crimson')
plt.title('Emergency Funds')

plt.subplot(2, 2, 4)
ax = sns.histplot(df['savings'], kde=True)
ax.lines[0].set_color('crimson')
plt.title('Savings')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.subplot(2, 2, 1)
sns.boxplot(x=df['income'])
plt.title('Income')

plt.subplot(2, 2, 2)
sns.boxplot(x=df['expense'])
plt.title('Expenses')

plt.subplot(2, 2, 3)
sns.boxplot(x=df['emergency_fund'])
plt.title('Emergency Funds')

plt.subplot(2, 2, 4)
sns.boxplot(x=df['savings'])
plt.title('Savings')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.scatterplot(x=df['income'], y=df['savings'])
plt.title('Income vs. Savings')

plt.subplot(1, 2, 2)
sns.scatterplot(x=df['expense'], y=df['savings'])
plt.title('Expenses vs. Savings')


plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(x=df['income'], y =df['expense'], hue=df['savings'])
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.pairplot(df, vars=['income', 'expense', 'emergency_fund', 'savings'])
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.regplot(x=df['income'], y=df['savings'], line_kws={"color": "red"})
plt.title('Income vs. Savings')

plt.subplot(1, 2, 2)
sns.regplot(x=df['expense'], y=df['savings'], scatter_kws={"color": "black"}, line_kws={"color": "red"})
plt.title('Expenses vs. Savings')

plt.tight_layout()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

x = df['income']
y = df['expense']
z = df['savings']

ax.scatter(x, y, z)

ax.set_xlabel('Income')
ax.set_ylabel('Expenses')
ax.set_zlabel('Savings')

plt.show()
```

```python

plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
sns.regplot(x=df['expense'], y=df['savings'], color='black', scatter=False)
plt.title('Correlation Matrix')
plt.show()
```
